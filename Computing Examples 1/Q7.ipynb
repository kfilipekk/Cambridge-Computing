```python
import numpy as np
import matplotlib.pyplot as plt

dt = 0.01
T = np.arange(0,20,dt)
m = 1
k = 40
w = np.sqrt(k / m)

x = [0] * len(T) #use numpy
x[0],x[1] = 0.01,0.01
F = 0

#Compute the numerical solution
for n in range(2, len(T)):
    x[n] = (dt**2)*(F-k*x[n-1]) + 2*x[n-1] - x[n-2]

#Compute the analytical solution
x_anly = 0.01 * np.cos(w * T)

plt.figure(figsize=(8, 4))
plt.plot(T, x, label="Numerical", linestyle="dashed")
plt.plot(T, x_anly, label="Analytical")
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
plt.title("Mass-Spring System")
plt.show()

```

```python
def friction_force(v):
    return -0.025 if v > 0 else 0.025 if v < 0 else 0.0

xf = np.zeros(len(T))
xf[0],xf[1] = 0.01,0.01

for n in range(2, len(T)):
    v = (xf[n-1] - xf[n-2]) / dt  #Approximate velocity
    xf[n] = (dt**2)*(friction_force(v)-k*xf[n-1]) + 2*xf[n-1] - xf[n-2]


plt.figure(figsize=(8, 4))
plt.plot(T, xf, label='Numerical (with friction)', linestyle='dashed')
plt.plot(T, x_anly, label="Analytical (no friction)")
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
plt.title("Mass-Spring System with Friction")
plt.show()
```
